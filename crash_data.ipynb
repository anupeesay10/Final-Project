# Pittburgh neighborhood crash data

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

allegheny_crash_data = pd.read_csv("2015allegheny.csv", sep=",", low_memory=False)
beaver_crash_data = pd.read_csv("2015beaver.csv", sep=",", low_memory=False)
butler_crash_data = pd.read_csv("2015butler.csv", sep=",", low_memory=False)
washington_crash_data = pd.read_csv("2015washington.csv", sep=",", low_memory=False)
westmoreland_crash_data = pd.read_csv("2015westmoreland.csv", sep=",", low_memory=False)
```

## Total crash count

```python
pd.Series([allegheny_crash_data['DISTRICT'].count(),beaver_crash_data['DISTRICT'].count(),butler_crash_data['DISTRICT'].count(),washington_crash_data['DISTRICT'].count(),westmoreland_crash_data['DISTRICT'].count()],
          index=["allegheny", "beaver", "butler", "washington", "westmoreland"]).plot.bar()
```

## Pie chart for each county

### Allegheny

```python
dataframe = pd.DataFrame({'Name': ['Non collision', ' Rear-end', 'Head-on',  'Rear-to-rear', 'Angle', 'Sideswipe Same Direction', "Sideswipe Opposide", "Hit fixed object", "Hit pedestrian", "unknown"],
                          'votes_of_each_class': allegheny_crash_data.groupby("COLLISION_TYPE").count()['DISTRICT']}) 
dataframe.groupby(['Name']).sum().plot( kind='pie', y='votes_of_each_class', autopct='%1.0f%%', figsize=(10,10)) 
```
